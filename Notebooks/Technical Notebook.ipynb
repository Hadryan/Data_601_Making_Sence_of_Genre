## 2) Introduction 

Music genre is an incredibly difficult classification question as the lines that separate one style of music from another are never definitive. There is always going to be songs and artist that bend and twist genre, eventually shaping their work into wholly new and original genres.

The inspiration for this project came from another tool that used the spotify web API. [Every Noise](http://everynoise.com/) is an ongoing project that is logging and visualizing the spotify genre space. As of November 2020 there are 5,068 unique genres within the Spotify track library, and the number is always growing. The tool provides a wonderfully exciting way to explore both mainstream and fringe genres and builds a playlist for each and every genre it logs.  [This Link](https://open.spotify.com/playlist/69fEt9DN5r4JQATi52sRtq?si=_xSevsMHTh6QEOnPmQdAGw) will take you to a playlist generate by the Every Noise Project, where every genre has one song; over 350 hours of music.

#### Spotify Data

For the scope of this propject we will be looking a only thirteen genres. 

[Spotipy](https://github.com/plamere/spotipy) is a python wrapper for the [Spotify Web API](https://developer.spotify.com/documentation/web-api/).  From this Using Every Noise, a large playlist (of up to 1300 songs) for each genre was, quiered in the and saved to a large data frame. The folloing features were collected and saved to the spotify_df.csv 

- track_id : a spotify primary key; unique for each track
- artist: name of artist
- album: name of album
- trackName: title of track
- acousticness: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
- danceability: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. 
- duration_ms: The duration of the track in milliseconds.
- energy: a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. 
- instrumentalness: Predicts whether a track contains no vocals. “Ooh” and “aah” sounds are treated as instrumental in this context.
- key: The estimated overall key of the track. Integers map to pitches using standard [Pitch Class notation](https://en.wikipedia.org/wiki/Pitch_class) 
- liveness: Detects the presence of an audience in the recording.
- loudness: The overall loudness of a track in decibels (dB). 
- mode: indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
- speechiness: Speechiness detects the presence of spoken words in a track.
- tempo: BPM of track
- time_signature: An estimated overall time signature of a track. 
- valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. 
- category_id: The Spotify Category ID of the track 
- popularity: The value will be between 0 and 100, with 100 being the most popular. The artist’s popularity is calculated from the popularity of all the artist’s tracks.

## 5) Data Prep & Feature Engineering 

### Simplifying Genre 

This dataset contains a total of 13 different genres and many of them overlap and genres that are very similar to one another. (For example, Rap and Hip Hop). There are also genres that have very similar audio features among all the tracks within, such as EDM. These genres may pose a challenge in classifying and for the sake of this project will be combined into more simplistic labels. 

We will reduce it down to 7 genre catigories 

 - Pop and Dance_pop will labled as "Pop/Dance_pop"
 - Rap and Hip Hop will be labled as "Rap/Hip Hop"
 - Rock, Alternitive_rock, indie_rock, and indie_folk will be lable as "Rock/indie/alt"
 - Punk and Metal will be grouped together as "Punk/Metal"
 - Country, edm and jazz will remain as is
 

```python
#combining genres 
#'pop/dance_pop' 
spotify_df['category_id'] = np.where((spotify_df['category_id'] == 'pop')| (spotify_df['category_id'] == 'dance_pop'),'pop/dance_pop',spotify_df['category_id'])
#'rap/hiphop'
spotify_df['category_id'] = np.where((spotify_df['category_id'] == 'rap')| (spotify_df['category_id'] == 'hip_hop'),'rap/hiphop',spotify_df['category_id'])
#'rock/alt/indie'
spotify_df['category_id'] = np.where((spotify_df['category_id'] == 'rock')| (spotify_df['category_id'] == 'alternative_rock')| (spotify_df['category_id'] == 'indie_rock') | (spotify_df['category_id'] == 'indie_folk'),
                                     'rock/indie/alt', spotify_df['category_id'])
#'punk/metal'
spotify_df['category_id'] = np.where((spotify_df['category_id'] == 'punk')| (spotify_df['category_id'] == 'metal'),'punk/metal',spotify_df['category_id'])
```

```python
# How many tracks per genre now that they have been simplfyied 
spotify_df.category_id.value_counts()
```

### Selecting our Feature Data 

Using what we learned during EDA, we can select the following audio features to be used in the model building process 

 -  Because Loudness and Energy were highly correlated, and there was minimal variation in loudness across the different genres, it will be dropped
 

```python
spotify_features_list = ["acousticness","danceability","duration_ms","energy","instrumentalness","key","liveness","mode","speechiness","tempo","time_signature","valence","popularity"]
spotify_features_X  = spotify_df[spotify_features_list].values
spotify_targets_y = spotify_df['category_id'].values
```

### Spliting 

 - In order to evaluate the effectiveness of the models, we will split the data into a training and test set. The test set will be used to evaluate the trained models on new data.
 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#set Random seed
rnd_seed = 48123412
```

```python
#spliting the data
X_train, X_test, y_train, y_test = train_test_split(spotify_features_X, spotify_targets_y, test_size = 0.3, random_state =rnd_seed, stratify =spotify_targets_y)
```

### Normalizing 

- To avoid having features with a large range of values introducing bias into the models, we must normalize the data. This will calculate the z score for each feature on a standard normal curve (mean = 0, standard deviation = 1)


```python
#scalling the data
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

### Testing The fit 

 - We will test how the data fits as a baseline before doing any feature engineering 

```python
#Logistic Regression on only scaled training data
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(penalty = 'none', random_state=rnd_seed,max_iter=3000)
logreg.fit(X_train_sc, y_train)
logreg.score(X_test_sc, y_test)
```

Our first model is better than a coin flip ! 

### Feature Engineering

#### Principal Componet Analysis

 - The current dataset has 13 different features, among thoes features there are differnt interactions and correlations.
 -  To improve the accuracy of our classifier, we could reduse the dimentiaonlity of our dataset
 - Redusing the dimentions can also improve the run time preformance of some more advancsed models  
 -  Principal Component Analysis or PCA can look at relationships and patterns and project the data to a lower dimensionality


```python
from sklearn.decomposition import PCA

pca = PCA().fit(X_train_sc)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.axhline(y=0.99, linestyle='--')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

For PCA a good heuristic for knowing how many components to include is to look for the n needed  to reach 90% of the variance. Looking at the cumulative variance plot above, it looks like we will need at least 9 components.

```python
componets = 12
# fiting test and train data to pca
pca = PCA(componets, random_state=rnd_seed)
pca.fit(X_train_sc)
X_train_pca = pca.transform(X_train_sc)
X_test_pca = pca.transform(X_test_sc)
```

Lets test to see if PCA would be useful in this problem, by comparing the logistic regression model score trained on only scaled data and data that has had its dimensions reduced by PCA

```python
#Logistic Regression on PCA redused training data
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(penalty = 'none', random_state=rnd_seed,max_iter=3000)
logreg.fit(X_train_pca, y_train)
logreg.score(X_test_pca, y_test)
```

There doesn't seam to be a significant difference between the scores of each so we will proceed without using the PCA data and Continue with the scaled data.

## 6) Modeling
