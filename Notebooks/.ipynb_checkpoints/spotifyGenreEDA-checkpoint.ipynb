### Answering Questions About the Dataset

```python
# What (and how many) Genres are in this dataset 
genre_list = ['pop','dance_pop','rap','rock','hip_hop','country','edm','indie_pop','indie_rock','punk','metal',
              'alternative_rock','jazz']
len(genre_list)
```

```python
# How many unique tracks?
spotify_df.track_id.nunique()
```

```python
# How many tracks per genre 
spotify_df.category_id.value_counts()
```

```python
# What Artists show up in multiple genres within the dataset
spotify_df.groupby('category_id')
```

####  Key Observations 


 - Not all genres have the same number of tracks, could affect any classifier if this not taken into account when building the training and test data sets
 - There are some songs that show up in multiple genres, this could be interesting to test how a classifier handles these tracks


```python
# How many unique artist?
spotify_df.artist.nunique()
```

```python
#What artists has the most tracks in the dataset?
spotify_df.artist.value_counts()[:50]
```

```python
# What track titles are the most frequent in the dataset?
spotify_df.trackName.value_counts()[:15]
```

```python
# What are the mean values of each genre
spotify_df.groupby('category_id').mean()
```

#### Key Observations 

 - Pop is the most popular genre, followed by Rap and dance pop
 - Rap and Hip Hop are the most danceable genres, and Metal is the least
